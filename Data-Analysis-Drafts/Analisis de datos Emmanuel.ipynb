# Migración de datos

```python
#Mostrar cantidad de datos en cada dataframe
len(order_reviews), len(order_payments), len(orders), len(order_items)
```

```python
#Juntar dataframes
df_order = pd.merge(left = order_items, right = orders,
                   how = "inner",
                   left_on = ["order_id"],
                   right_on = ["order_id"])
df_order.head()
```

```python
df_order = pd.merge(left = df_order, right = order_payments,
                   how = "inner",
                   left_on = ["order_id"],
                   right_on = ["order_id"])
df_order.head()
```

```python
df_order = pd.merge(left = df_order, right = order_reviews,
                   how = "inner",
                   left_on = ["order_id"],
                   right_on = ["order_id"])
df_order.head()
```

```python
df_order = pd.merge(left = df_order, right = order_reviews,
                   how = "inner",
                   left_on = ["order_id"],
                   right_on = ["order_id"])
df_order.head()
```

```python
df_order_customer = pd.merge(left = df_order, right = customers,
                   how = "inner",
                   left_on = ["customer_id"],
                   right_on = ["customer_id"])
df_order_customer.head()
```

```python
df_order_customer_products = pd.merge(left = df_order_customer, right = products,
                   how = "inner",
                   left_on = ["product_id"],
                   right_on = ["product_id"])
df_order_customer_products.head()
```

```python
df_order_customer_products_sellers = pd.merge(left = df_order_customer_products, right = sellers,
                   how = "inner",
                   left_on = ["seller_id"],
                   right_on = ["seller_id"])
df_order_customer_products_sellers.head()
```

```python
#Dataframe completo de order, customer, products y sellers
df_order_customer_products_sellers.columns
```

```python
#Grabar el dataframe a un csv
df_order_customer_products_sellers.to_csv('alldata.csv')
```

# Revisión visual de datos

```python
#Agrupar por seller_id y contar por el numero de ordenes generadas
sellersbest = df_order_customer_products_sellers.groupby(['seller_id']).order_id.count()
sellersbest
```

```python
#Visualización de los datos
sellersbest = df_order_customer_products_sellers.groupby(['seller_id']).order_id.count()
sellersbest = pd.DataFrame(sellersbest)
sellersbest.reset_index(inplace = True)
sellersbest = sellersbest.sort_values(["order_id"], ascending=False)
#solo mostrar 15 datos
sb15 = sellersbest.head(15)
plt.figure(figsize=(8,14))
sns.barplot(x=sb15.order_id, y=sb15.seller_id, data=sb15)
sellersbest.head()
```

```python
#Vizualicacion de agrupación por categorias de productos, la mediana por el precio del producto
prodprice = df_order_customer_products_sellers.groupby(['product_category_name']).price.median()
prodprice = pd.DataFrame(prodprice)
prodprice.reset_index(inplace = True)
prodprice = prodprice.sort_values(["price"], ascending=False)
#prodprice.plot(x = "price", legend = False, xerr = "std")
plt.figure(figsize=(8,14))
sns.barplot(x=prodprice.price, y=prodprice.product_category_name, data=prodprice)
prodprice.head()
```

```python
#Vizualicacion de agrupación por categorias de productos, el precio máximo del producto
prodmax = df_order_customer_products_sellers.groupby(['product_category_name']).price.max()
prodmax = pd.DataFrame(prodmax)
```
